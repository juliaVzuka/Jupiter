4.sin(а*x)=0
при условии: 0.01<a<0.02, 100<х<500.
Решение:
x = Pi*n / a, n = 0,1,2,...
Как видно из графика ниже, только при n = 1,2 и 3 существуют a из интервала 0.01<a<0.02, такие что 100<х<500
Итого x = Pi/a, при 0.01<a<0.02; x = 2Pi/a, при a1<a<0.02; x = 3Pi/a, при a2<a<0.02
a1 = 0.0126 и a2 = 0.0188 вычислены в программе.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
for n in range(1, 4):
   a = np.linspace(0.01, 0.02, 1000)
   x = np.pi * n / a
   plt.ylim(100, 500)
   plt.plot(a, x, label = f"n = {n}")
   plt.legend(loc = 'lower left')
   plt.xlabel('a')
   plt.ylabel('x') 
a1 = 2*np.pi / 500
a2 = 3*np.pi / 500
print(a1,a2)
```
